```
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten,BatchNormalization, Dropout, MaxPooling2D
import cv2
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
sheep = []
image = cv2.imread("/content/Delaine-ewe.jpg")
image = cv2.resize(image,(256,256))
sheep.append(image)
```

```
sheep = np.array(sheep)
```

```
sheep.shape
```

```
datagen = ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True)
```

```
i = 0
for batch in datagen.flow(sheep, batch_size = 1, save_to_dir='/content/sheep',save_format='jpeg',save_prefix='sheep'):
  i = i + 1
  if(i > 50):
    break
```

```
car = []
image = cv2.imread("/content/SONATA-hero-option1-764A5360-edit.jpg")
image = cv2.resize(image,(256,256))
car.append(image)
```

```
car = np.array(car)
```

```
i = 0
for batch in datagen.flow(car, batch_size=1, save_to_dir='/content/car', save_format = 'jpeg', save_prefix = 'car'):
  i = i+1
  if(i >50):
    break
```

```
X = []
Y = []
images_name = os.listdir("/content/car")
for i in range(len(images_name)):
  try:
    image = cv2.imread("/content/car/" + images_name[i])
    image = cv2.resize(image,(256,256))
    X.append(image)
    Y.append(1)
  except:
    pass
```

```
images_name = os.listdir("/content/sheep")
for i in range(len(images_name)):
  try:
    image = cv2.imread("/content/sheep/" + images_name[i])
    image = cv2.resize(image,(256,256))
    X.append(image)
    Y.append(0)
  except:
    pass
```

```
X = np.array(X)
X.shape
```

```
Y = np.array(Y)
```

```
model = Sequential()
```

```
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
```

```
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
model.fit(X,Y,epochs=10,batch_size=10)
```

```
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
```

```
inputs = Input(shape= (256,256,3))
output1 = Conv2D(32,(3,3),activation='relu')(inputs)
output2 = MaxPooling2D(pool_size=(2,2))(output1)
output3 = Conv2D(32,(3,3),activation='relu')(output2)
output4 = MaxPooling2D(pool_size=(2,2))(output3)
output5 = Flatten()(output4)
output6 = Dense(64, activation='relu')(output5)
output7 = Dense(128, activation='relu')(output6)
output8 = Dense(256, activation='relu')(output7)
outputs = Dense(1,activation='sigmoid')(output8)
```

```
model = Model(inputs = inputs, outputs = outputs)
```

```
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
model.fit(X,Y,epochs= 10, batch_size=10)
```
