Images generator

```
datagen = ImageDataGenerator(
    featurewise_center=True, 
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True)
```

Extraction of images with different parameters and saving in directory

```
i = 0
for batch in datagen.flow(car, batch_size=1, 
                          save_to_dir='/content/drive/My Drive/Data Science/Data Augmentation/tesla cars', 
                          save_prefix='car', 
                          save_format='jpeg'):
  i+=1
  if(i>60):
    break
```

Creating Features and Labels of Cars

```
X = []
y = []

images_name = os.listdir("/content/drive/My Drive/Data Science/Data Augmentation/tesla cars/")
for i in range(len(images_name)):
  image = cv2.imread("/content/drive/My Drive/Data Science/Data Augmentation/tesla cars/" + images_name[i])
  image = cv2.resize(image,(256,256))
  X.append(image)
  y.append(1)
```

```
y[0]
```

```
X[0]
```

Creating Features and Labels of Trucks

```
truck_names = os.listdir('/content/drive/My Drive/Data Science/Data Augmentation/trucks/')

for i in range(len(truck_names)):
  try:
    image = cv2.imread('/content/drive/My Drive/Data Science/Data Augmentation/trucks/' + truck_names[i])
    image = cv2.resize(image,(256,256))
    X.append(image)
    y.append(0)
  except:
    pass
```

```
y[0]
```

```
X[0]
```

Conversion of Features and Labels in array

```
X = np.array(X)
y = np.array(y)
X.shape, y.shape
```

Normalizing Features

```
X = X/255
```

Building model

```
model = Sequential()
```

Building Convolution Layers

```
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

Flattening of Input

```
model.add(Flatten())
```

BUilding Neural Network

```
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid'))
```

```
model.summary()
```

```
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
model.fit(X, y, batch_size=100, epochs=10)
```
