```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```
data = pd.read_csv('/content/drive/My Drive/Data Science/Salary_Data.csv')
data.head()
```

```
data.info()
```

```
data.describe()
```

```
plt.scatter(data['YearsExperience'],data['Salary'], c='red')
```

```
X = data.iloc[:,0]
y = data.iloc[:,1]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```
X_train = np.array(X_train)
X_train = X_train.reshape(-1,1)
X_train.shape
```

```
X_test = np.array(X_test)
X_test = X_test.reshape(-1,1)
X_test.shape
```

```
y_train = np.array(y_train)
y_train = y_train.reshape(-1,1)
y_train.shape
```

```
y_test = np.array(y_test)
```

```
model = LinearRegression()
model.fit(X_train,y_train)
```

```
y_pred = model.predict(X_test)
y_pred
```

```
y_pred = y_pred.reshape(-1)
y_pred
```

```
y_pred = pd.DataFrame(y_pred, columns=['pred_y'])
y_pred
```

```
y_test = pd.DataFrame(y_test, columns=['test_y'])
y_test
```

```
pred_val = pd.concat([y_test, y_pred], axis=1)
pred_val
```
