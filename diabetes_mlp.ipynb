Import libraries


```
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from sklearn.model_selection import train_test_split

```

Load the data

```
dataset = pd.read_csv("/content/drive/My Drive/Data Science/diabetes.csv")
```

```
dataset.head()
```

```
dataset.groupby('Outcome').count()
```

```
dataset.describe()
```

```
x = dataset.iloc[:,:-1] #featues
y = dataset.iloc[:,-1] #labels
```

```
x.shape
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```
x_train
```

Building Neural network


```
model = Sequential()
```

```
model.add(Dense(32, input_dim = 8, activation=None)) #8 bcoz 8 columns
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu')) #Hidden layer
model.add(Dense(1, activation='sigmoid')) #output layer
```

```
model.summary()
```

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(x_train, y_train, epochs=20, batch= 50)
```
